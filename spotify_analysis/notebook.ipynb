```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import kagglehub
import os
import seaborn as sns
pd.set_option("display.max_rows", 20)
pd.set_option("display.max_columns", 30)
```

```python
path = kagglehub.dataset_download("gauthamvijayaraj/spotify-tracks-dataset-updated-every-week")

print("Path to dataset files:", path)
```

```python
print(os.listdir(path))
```

```python
df = pd.read_csv(path + "/spotify_tracks.csv")
```

```python
features = df.select_dtypes(include=["number"]).columns.tolist()
print(features)
```

```python
df[features].corr()["popularity"].sort_values(ascending=False)
```

```python
hits = df[df["popularity"]>=70]
non_hits = df[df["popularity"]<70]
```

```python
comparison = pd.DataFrame({
    "hits": hits[features].mean(),
    "non_hits": non_hits[features].mean()
})

comparison
```

```python
diff = (comparison["hits"] - comparison["non_hits"]) / comparison["non_hits"]
diff.drop(["popularity", "year", "duration_ms", ]).sort_values().plot(kind="barh", title="Difference in means of features between hits and non-hits")
```

```python
df["status"].value_counts()
```

```python
'''
common_norm=False means that the area under the curve for each category will be normalized separately,
 so that the total area for each category is equal to 1. 
This allows for a better comparison of the distributions of the two categories, regardless of their sample sizes.
'''
df["status"] = np.where(df["popularity"] >= 70, "hit", "non-hit")
sns.kdeplot(data=df, x="instrumentalness", hue="status", fill=True, common_norm=False)
#sns.kdeplot(data=df, x="energy", hue=(df["popularity"]>=70), fill=True)
```

```python
sns.violinplot(
    data=df,
    x="status",
    y="instrumentalness"
)
```

```python
sns.histplot(
    data=df,
    x="instrumentalness",
    hue="status",
    stat="density",
    element="step",
    common_norm=False,
    bins=10
)
```
